```python
import time
import os
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import re
from datetime import datetime

def diagnosticar_extracao_tjpr():
    """
    Script de diagnóstico para identificar problemas na extração de dados do TJPR
    """
    # Configurando o Selenium
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--window-size=1920,1080')

    # Verificar o ambiente
    if os.path.exists('/usr/bin/chromium-browser'):
        chrome_options.binary_location = '/usr/bin/chromium-browser'

    # Criando pasta para diagnóstico
    output_dir = f"diagnostico_tjpr_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    os.makedirs(output_dir, exist_ok=True)
    print(f"Pasta de diagnóstico criada: {output_dir}")

    # Iniciando o Chrome
    driver = webdriver.Chrome(options=chrome_options)
    print("Chrome inicializado com sucesso!")

    try:
        # Acessando a página
        url = 'https://portal.tjpr.jus.br/jurisprudencia/'
        driver.get(url)
        print(f"Acessando URL: {url}")

        # Tirando screenshot inicial
        driver.save_screenshot(f"{output_dir}/01_pagina_inicial.png")

        # Aguardando o formulário carregar
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "pesquisaForm"))
        )
        print("Formulário de pesquisa encontrado.")

        # Preenchendo critérios exatos de pesquisa
        campo_pesquisa = driver.find_element(By.ID, "criterioPesquisa")
        campo_pesquisa.clear()
        campo_pesquisa.send_keys('"remessa necessária"')

        botao_ou = driver.find_element(By.ID, "btOu")
        botao_ou.click()

        campo_pesquisa.send_keys('"reexame necessário"')

        campo_data_inicial = driver.find_element(By.ID, "dataJulgamentoInicio")
        campo_data_inicial.clear()
        campo_data_inicial.send_keys("01/01/2024")

        campo_data_final = driver.find_element(By.ID, "dataJulgamentoFim")
        campo_data_final.clear()
        campo_data_final.send_keys("31/12/2024")

        dropdown_ementa = driver.find_element(By.ID, "idLocalPesquisa")
        select_ementa = Select(dropdown_ementa)
        select_ementa.select_by_visible_text("EMENTA")

        dropdown_segredo = driver.find_element(By.ID, "idSegredoJustica")
        select_segredo = Select(dropdown_segredo)
        select_segredo.select_by_visible_text("EXCLUIR")

        # Screenshot do formulário preenchido
        driver.save_screenshot(f"{output_dir}/02_formulario_preenchido.png")

        # Clicando no botão de pesquisa
        botao_pesquisa = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.NAME, "iniciar"))
        )
        botao_pesquisa.click()
        print("Botão de pesquisa clicado.")

        # Aguardar carregamento com tempo ampliado
        print("Aguardando carregamento dos resultados (tempo ampliado)...")
        time.sleep(15)  # Tempo ampliado para garantir carregamento

        # Screenshot da página de resultados
        driver.save_screenshot(f"{output_dir}/03_pagina_resultados.png")

        # Salvar código-fonte da página
        html = driver.page_source
        with open(f"{output_dir}/pagina_resultados.html", "w", encoding="utf-8") as f:
            f.write(html)
        print("HTML da página de resultados salvo para análise")

        # Analisar a estrutura HTML para diagnosticar
        soup = BeautifulSoup(html, 'html.parser')

        # 1. Verificar número de registros
        div_navLeft = soup.find('div', class_='navLeft')
        if div_navLeft:
            print(f"Conteúdo da div navLeft: {div_navLeft.text.strip()}")

            match = re.search(r'(\d+)\s+registro\(s\)\s+encontrado\(s\)', div_navLeft.text)
            if match:
                total_registros = int(match.group(1))
                print(f"Total de registros encontrados: {total_registros}")
            else:
                print("Não foi possível extrair o número de registros")
        else:
            print("Div navLeft não foi encontrada")

        # 2. Procurar elementos com a classe 'registro'
        registros = soup.find_all('div', class_='registro')
        print(f"Número de elementos com a classe 'registro': {len(registros)}")

        if len(registros) == 0:
            print("PROBLEMA DETECTADO: Nenhum elemento com classe 'registro' encontrado")

            # Listar todas as classes div na página para diagnóstico
            print("\nListando todas as classes div na página:")
            todas_classes = set()
            for div in soup.find_all('div'):
                if 'class' in div.attrs:
                    for classe in div['class']:
                        todas_classes.add(classe)

            print(f"Classes encontradas: {sorted(todas_classes)}")

            # Tentar encontrar elementos que possam conter os registros
            print("\nProcurando por possíveis elementos de resultado:")
            # Lista de possíveis classes/ids para procurar
            candidatos = ['resultado', 'result', 'registros', 'records', 'item', 'acordao']

            for candidato in candidatos:
                elementos_por_classe = soup.find_all(class_=re.compile(candidato, re.IGNORECASE))
                elementos_por_id = soup.find_all(id=re.compile(candidato, re.IGNORECASE))

                if elementos_por_classe:
                    print(f"Encontrados {len(elementos_por_classe)} elementos com classe contendo '{candidato}'")
                    for elem in elementos_por_classe[:3]:  # Mostrar apenas os 3 primeiros
                        print(f"  - Tag: {elem.name}, Classes: {elem.get('class')}")

                if elementos_por_id:
                    print(f"Encontrados {len(elementos_por_id)} elementos com ID contendo '{candidato}'")
                    for elem in elementos_por_id[:3]:
                        print(f"  - Tag: {elem.name}, ID: {elem.get('id')}")

        # 3. Verificar navegação
        links_navegacao = soup.find_all('a', text=re.compile(r'Próxima|Next|Seguinte'))
        print(f"\nNúmero de links de 'Próxima página' encontrados: {len(links_navegacao)}")

        if len(links_navegacao) == 0:
            print("PROBLEMA DETECTADO: Nenhum link de 'Próxima página' encontrado")

            # Listar todos os links na página
            print("\nListando todos os links na página:")
            todos_links = soup.find_all('a')
            print(f"Total de links: {len(todos_links)}")

            # Mostrar os textos dos links para identificar o de navegação
            textos_links = [link.text.strip() for link in todos_links if link.text.strip()]
            print(f"Textos dos links: {textos_links[:20]}")  # Mostrar até 20 textos

        # Tentar clicar no link de próxima página e verificar comportamento
        try:
            link_proxima = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(text(), 'Próxima')]"))
            )
            print("\nLink 'Próxima' encontrado no DOM via Selenium. Tentando clicar...")

            # Screenshot antes de clicar
            driver.save_screenshot(f"{output_dir}/04_antes_navegacao.png")

            # Tentar clicar no link
            driver.execute_script("arguments[0].scrollIntoView();", link_proxima)
            time.sleep(1)
            link_proxima.click()

            print("Clique no link 'Próxima' executado. Aguardando carregamento...")
            time.sleep(10)

            # Screenshot após clicar
            driver.save_screenshot(f"{output_dir}/05_apos_navegacao.png")

            # Verificar se houve mudança na URL ou conteúdo
            nova_url = driver.current_url
            print(f"URL após clique: {nova_url}")

            # Salvar HTML após navegação
            with open(f"{output_dir}/pagina_apos_navegacao.html", "w", encoding="utf-8") as f:
                f.write(driver.page_source)

        except Exception as e:
            print(f"Erro ao tentar navegar para próxima página: {e}")

        # Finalizando diagnóstico
        print("\nDiagnóstico concluído. Verifique a pasta de saída para análise detalhada.")
        return True

    except Exception as e:
        print(f"Erro durante o diagnóstico: {e}")
        driver.save_screenshot(f"{output_dir}/erro_diagnostico.png")
        return False

    finally:
        driver.quit()
        print("Navegador fechado.")

if __name__ == "__main__":
    print("Iniciando diagnóstico da extração do TJPR...")
    diagnosticar_extracao_tjpr()
```

```python
import time
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from bs4 import BeautifulSoup
import re
from datetime import datetime

def extrair_jurisprudencia_tjpr():
    """
    Script para extrair jurisprudência do TJPR usando a estrutura correta da página
    """
    # Configurando o Selenium
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--window-size=1920,1080')

    # Verificar o ambiente
    if os.path.exists('/usr/bin/chromium-browser'):
        chrome_options.binary_location = '/usr/bin/chromium-browser'

    # Criando pasta para resultados
    output_dir = f"extracao_tjpr_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    os.makedirs(output_dir, exist_ok=True)
    print(f"Pasta de saída criada: {output_dir}")

    # Iniciando o Chrome
    driver = webdriver.Chrome(options=chrome_options)
    print("Chrome inicializado com sucesso!")

    # Lista para armazenar todos os registros
    todos_registros = []

    try:
        # Acessando a página
        url = 'https://portal.tjpr.jus.br/jurisprudencia/'
        driver.get(url)
        print(f"Acessando URL: {url}")

        # Aguardando o formulário carregar
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "pesquisaForm"))
        )
        print("Formulário de pesquisa encontrado.")

        # Preenchendo critérios exatos de pesquisa
        campo_pesquisa = driver.find_element(By.ID, "criterioPesquisa")
        campo_pesquisa.clear()
        campo_pesquisa.send_keys('"remessa necessária"')

        botao_ou = driver.find_element(By.ID, "btOu")
        botao_ou.click()

        campo_pesquisa.send_keys('"reexame necessário"')

        campo_data_inicial = driver.find_element(By.ID, "dataJulgamentoInicio")
        campo_data_inicial.clear()
        campo_data_inicial.send_keys("01/01/2024")

        campo_data_final = driver.find_element(By.ID, "dataJulgamentoFim")
        campo_data_final.clear()
        campo_data_final.send_keys("31/12/2024")

        dropdown_ementa = driver.find_element(By.ID, "idLocalPesquisa")
        select_ementa = Select(dropdown_ementa)
        select_ementa.select_by_visible_text("EMENTA")

        dropdown_segredo = driver.find_element(By.ID, "idSegredoJustica")
        select_segredo = Select(dropdown_segredo)
        select_segredo.select_by_visible_text("EXCLUIR")

        # Screenshot do formulário preenchido
        driver.save_screenshot(f"{output_dir}/01_formulario_preenchido.png")

        # Clicando no botão de pesquisa
        botao_pesquisa = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.NAME, "iniciar"))
        )
        botao_pesquisa.click()
        print("Botão de pesquisa clicado.")

        # Aguardar carregamento com tempo ampliado
        print("Aguardando carregamento dos resultados...")
        time.sleep(7)

        # Screenshot da página de resultados
        driver.save_screenshot(f"{output_dir}/02_pagina_resultados.png")

        # Verificar o número total de registros
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # Extrair o número total de registros
        div_navLeft = soup.find('div', class_='navLeft')
        if div_navLeft:
            texto = div_navLeft.text
            match = re.search(r'(\d+)\s+registro\(s\)\s+encontrado\(s\)', texto)
            if match:
                total_registros = int(match.group(1))
                # Calculando o número de páginas (10 registros por página)
                total_paginas = (total_registros + 9) // 10  # Arredondando para cima
                print(f"Total de registros encontrados: {total_registros}")
                print(f"Total de páginas a navegar: {total_paginas}")
        else:
            print("Não foi possível encontrar o número total de registros.")
            return []

        # Salvar HTML da primeira página para análise
        with open(f"{output_dir}/pagina_resultados_1.html", "w", encoding="utf-8") as f:
            f.write(html)

        pagina_atual = 1

        while pagina_atual <= total_paginas:
            print(f"\nProcessando página {pagina_atual} de {total_paginas}...")

            # Extrair os dados da página atual
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')

            # Verificar se estamos na página correta
            div_navLeft = soup.find('div', class_='navLeft')
            if div_navLeft:
                texto = div_navLeft.text
                exibindo_match = re.search(r'exibindo de (\d+) até (\d+)', texto)
                if exibindo_match:
                    exibindo_de = int(exibindo_match.group(1))
                    exibindo_ate = int(exibindo_match.group(2))
                    registros_esperados_na_pagina = min(10, total_registros - (pagina_atual - 1) * 10)

                    # Verificar se os números da paginação são coerentes
                    if exibindo_de != ((pagina_atual - 1) * 10) + 1:
                        print(f"ATENÇÃO: Possível problema na navegação. Exibindo de {exibindo_de}, mas esperado {((pagina_atual - 1) * 10) + 1}")

                    # Salvar HTML para diagnóstico a cada 50 páginas
                    if pagina_atual % 50 == 0 or pagina_atual <= 3:
                        with open(f"{output_dir}/pagina_resultados_{pagina_atual}.html", "w", encoding="utf-8") as f:
                            f.write(html)

            # Na estrutura atual, os registros estão em <tr> com classes 'even' ou 'odd' dentro da tabela de resultados
            tabela_resultados = soup.find('table', class_='resultTable jurisprudencia')

            # Confirmar que encontrou a tabela de resultados
            if not tabela_resultados:
                print(f"ERRO: Tabela de resultados não encontrada na página {pagina_atual}")
                # Tirar screenshot do erro
                driver.save_screenshot(f"{output_dir}/erro_pagina_{pagina_atual}.png")
                # Continuar para a próxima página
                pagina_atual += 1
                continue

            # Encontrar todas as linhas de resultado (even e odd)
            linhas_resultados = tabela_resultados.find_all('tr', class_=re.compile(r'even|odd'))

            print(f"Encontrados {len(linhas_resultados)} registros na página {pagina_atual}")

            # Processar cada linha de resultado
            for idx, linha in enumerate(linhas_resultados, 1):
                try:
                    # Extrair dados do registro
                    registro = {}

                    # Coluna de dados do processo
                    coluna_dados = linha.find('td', class_='juris-tabela-dados')
                    if coluna_dados:
                        # Extrair número do processo
                        links_processo = coluna_dados.find_all('div', style='display: inline-block;')
                        if links_processo:
                            registro['numero_processo'] = links_processo[0].text.strip()

                        # Extrair tipo de decisão
                        decisao_elem = coluna_dados.find('font', class_=re.compile(r'decisao|acordao'))
                        if decisao_elem:
                            registro['tipo_decisao'] = decisao_elem.text.strip()

                        # Extrair relator
                        propriedades_div = coluna_dados.find('div', class_='juris-tabela-propriedades')
                        if propriedades_div:
                            texto_propriedades = propriedades_div.get_text()

                            # Extrair relator
                            relator_match = re.search(r'Relator:\s*(.*?)(?:\s*<br>|\s*Processo)', texto_propriedades, re.DOTALL)
                            if relator_match:
                                registro['relator'] = relator_match.group(1).strip()

                            # Extrair órgão julgador
                            orgao_match = re.search(r'Órgão Julgador:\s*(.*?)(?:<br>|Data)', texto_propriedades)
                            if orgao_match:
                                registro['orgao_julgador'] = orgao_match.group(1).strip()

                            # Extrair data de julgamento
                            data_match = re.search(r'Data Julgamento:\s*(\d{2}/\d{2}/\d{4})', texto_propriedades)
                            if data_match:
                                registro['data_julgamento'] = data_match.group(1).strip()

                    # Coluna da ementa
                    coluna_ementa = linha.find('td', class_='juris-tabela-ementa')
                    if coluna_ementa:
                        ementa_div = coluna_ementa.find('div', id=re.compile(r'ementa\d+'))
                        if ementa_div:
                            registro['ementa'] = ementa_div.get_text(separator=' ', strip=True)

                    # URL do processo
                    link_processo = linha.find('a', href=True)
                    if link_processo:
                        registro['url'] = 'https://portal.tjpr.jus.br' + link_processo['href']

                    # Adicionar dados de página
                    registro['pagina'] = pagina_atual
                    registro['posicao_na_pagina'] = idx

                    # Adicionar à lista de todos os registros
                    todos_registros.append(registro)
                    print(f"  Extraído registro {idx} da página {pagina_atual} - Processo: {registro.get('numero_processo', 'N/A')}")

                except Exception as e:
                    print(f"  Erro ao extrair registro {idx} da página {pagina_atual}: {e}")

            # Verificar número de registros extraídos nesta página
            registros_extraidos_pagina = sum(1 for r in todos_registros if r.get('pagina') == pagina_atual)
            print(f"Extraídos {registros_extraidos_pagina} registros na página {pagina_atual}")

            # Salvar progresso parcial a cada 10 páginas
            if pagina_atual % 10 == 0 or pagina_atual == total_paginas:
                arquivo_parcial = f"{output_dir}/resultados_parcial_ate_pagina_{pagina_atual}.csv"
                salvar_registros_csv(todos_registros, arquivo_parcial)
                print(f"Progresso salvo em: {arquivo_parcial}")

            # Se for a última página, sair do loop
            if pagina_atual == total_paginas:
                break

            # Navegar para a próxima página usando JavaScript
            try:
                # No TJPR, a navegação é feita via JavaScript.
                # Vamos usar o script que está no href do link de próxima página
                link_proxima = driver.find_element(By.XPATH, "//a[contains(@class, 'arrowNextOn')]")

                if link_proxima:
                    # Rolar até o elemento para garantir que está visível
                    driver.execute_script("arguments[0].scrollIntoView();", link_proxima)
                    time.sleep(1)

                    # Extrair e executar o JavaScript do link
                    onclick_script = link_proxima.get_attribute('href')
                    if onclick_script and 'javascript:' in onclick_script:
                        # Extrair apenas o código JS, removendo 'javascript:'
                        js_code = onclick_script.replace('javascript:', '')

                        # Executar o JavaScript
                        driver.execute_script(js_code)
                        print(f"Navegando para a página {pagina_atual + 1} via JavaScript...")
                    else:
                        # Se não tiver javascript no href, tentar clicar diretamente
                        link_proxima.click()
                        print(f"Navegando para a página {pagina_atual + 1} via clique...")

                    # Aguardar carregamento da nova página
                    time.sleep(5)

                    # Tirar screenshot a cada 50 páginas
                    if (pagina_atual + 1) % 50 == 0:
                        driver.save_screenshot(f"{output_dir}/navegacao_pagina_{pagina_atual+1}.png")

                    # Incrementar contador de página
                    pagina_atual += 1
                else:
                    print("Botão 'Próxima' não encontrado. Finalizando extração.")
                    break

            except Exception as e:
                print(f"Erro ao navegar para a próxima página: {e}")
                # Tentar uma abordagem alternativa usando outro seletor
                try:
                    # Tentar encontrar o link pelo texto
                    driver.execute_script("""
                        var links = document.getElementsByTagName('a');
                        for (var i = 0; i < links.length; i++) {
                            if (links[i].textContent.includes('Próxima')) {
                                links[i].click();
                                return true;
                            }
                        }
                        return false;
                    """)
                    print("Tentativa alternativa de navegação executada.")
                    time.sleep(5)
                    pagina_atual += 1
                except Exception as e2:
                    print(f"Falha na navegação alternativa: {e2}")
                    driver.save_screenshot(f"{output_dir}/erro_navegacao_pagina_{pagina_atual}.png")
                    # Se não conseguir navegar, tenta ir para a próxima página diretamente
                    try:
                        # Tentar ir para a URL da próxima página diretamente
                        proxima_pagina = pagina_atual + 1
                        script_navegacao = f"""
                            document.forms['pesquisaForm']['pageNumber'].value='{proxima_pagina}';
                            document.forms['pesquisaForm']['sortColumn'].value='processo_sDataJulgamento';
                            document.forms['pesquisaForm']['sortOrder'].value='DESC';
                            document.forms['pesquisaForm'].submit();
                        """
                        driver.execute_script(script_navegacao)
                        print(f"Navegação direta para página {proxima_pagina} tentada via script.")
                        time.sleep(5)
                        pagina_atual += 1
                    except Exception as e3:
                        print(f"Todas as tentativas de navegação falharam: {e3}")
                        break

        # Salvando todos os resultados no arquivo CSV final
        arquivo_final = f"{output_dir}/resultados_tjpr_completo.csv"
        salvar_registros_csv(todos_registros, arquivo_final)

        print(f"\nExtração concluída! Total de {len(todos_registros)} registros salvos em: {arquivo_final}")
        taxa_sucesso = len(todos_registros) / total_registros * 100 if total_registros > 0 else 0
        print(f"Taxa de sucesso: {taxa_sucesso:.2f}% dos registros extraídos.")

        return todos_registros

    except Exception as e:
        print(f"Erro global durante a execução: {e}")
        # Tenta salvar os dados que já foram extraídos
        if todos_registros:
            arquivo_emergencia = f"{output_dir}/resultados_emergencia.csv"
            salvar_registros_csv(todos_registros, arquivo_emergencia)
            print(f"Dados parciais salvos em: {arquivo_emergencia}")

        return todos_registros

    finally:
        # Fechando o navegador
        driver.quit()
        print("Navegador fechado.")

def salvar_registros_csv(registros, arquivo_saida):
    """
    Salva os registros extraídos em um arquivo CSV
    """
    try:
        if not registros:
            print("Nenhum registro para salvar.")
            return False

        # Converter para DataFrame
        df = pd.DataFrame(registros)

        # Definir a ordem das colunas principais (deixando as demais no final)
        colunas_principais = [
            'numero_processo', 'tipo_decisao', 'relator', 'orgao_julgador',
            'data_julgamento', 'ementa', 'url', 'pagina', 'posicao_na_pagina'
        ]

        # Colunas disponíveis no DataFrame
        todas_colunas = df.columns.tolist()

        # Reordenar: primeiro as colunas principais (que existem), depois as demais
        colunas_principais_existentes = [col for col in colunas_principais if col in todas_colunas]
        outras_colunas = [col for col in todas_colunas if col not in colunas_principais]

        colunas_ordenadas = colunas_principais_existentes + outras_colunas
        df = df[colunas_ordenadas]

        # Salvar como CSV com codificação UTF-8
        df.to_csv(arquivo_saida, index=False, encoding='utf-8-sig')

        return True

    except Exception as e:
        print(f"Erro ao salvar CSV: {e}")
        return False

def extrair_por_lotes(total_paginas, tamanho_lote=50):
    """
    Executa a extração em lotes de páginas para minimizar problemas de memória/tempo
    """
    todos_resultados = []
    diretorio_base = f"extracao_tjpr_lotes_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    os.makedirs(diretorio_base, exist_ok=True)

    # Número de lotes
    num_lotes = (total_paginas + tamanho_lote - 1) // tamanho_lote

    for lote in range(1, num_lotes + 1):
        pagina_inicial = (lote - 1) * tamanho_lote + 1
        pagina_final = min(lote * tamanho_lote, total_paginas)

        print(f"\n\nIniciando extração do lote {lote}/{num_lotes} (páginas {pagina_inicial} a {pagina_final})")

        # Implementar função específica para extrair um intervalo de páginas
        # Esta função seria semelhante à extrair_jurisprudencia_tjpr, mas com parâmetros para página inicial/final

        # Ao final, juntar todos os resultados
        arquivo_final = f"{diretorio_base}/resultados_completos.csv"
        # salvar_registros_csv(todos_resultados, arquivo_final)

    return arquivo_final

# Executando a extração
if __name__ == "__main__":
    print("Iniciando extração dos registros de jurisprudência do TJPR...")
    registros = extrair_jurisprudencia_tjpr()

    if registros:
        print(f"\nTotal de registros extraídos: {len(registros)}")
    else:
        print("Não foi possível extrair os registros.")
```

```python
import pandas as pd
import re

def mapear_tipo_decisao(tipo):
    """
    Mapeia o tipo de decisão de float para texto
    """
    mapeamento = {
        1.0: 'Decisão Monocrática',
        2.0: 'Acórdão',
        3.0: 'Decisão Monocrática',
        4.0: 'Acórdão',
        5.0: 'Acórdão'
    }
    return mapeamento.get(tipo, 'Não Especificado')

def extrair_orgao_julgador(relator):
    """
    Tenta extrair o órgão julgador do nome do relator
    """
    camaras = [
        '1ª Câmara Cível', '2ª Câmara Cível', '3ª Câmara Cível',
        '4ª Câmara Cível', '5ª Câmara Cível', '6ª Câmara Cível',
        '7ª Câmara Cível', '8ª Câmara Cível', '9ª Câmara Cível',
        '10ª Câmara Cível', '11ª Câmara Cível', '12ª Câmara Cível'
    ]

    # Procura por câmara no nome do relator
    for camara in camaras:
        if camara in relator:
            return camara

    return 'Não Especificado'

def extract_keywords(ementa):
    """
    Extrai palavras-chave da ementa com mais contexto
    """
    # Expressões regulares para capturar termos entre aspas e termos-chave
    keywords_pattern = r'"([^"]+)"'
    termos_chave = [
        'remessa necessária', 'reexame necessário',
        'ação civil', 'mandado de segurança',
        'auxílio', 'aposentadoria', 'benefício'
    ]

    # Extrair termos entre aspas
    keywords_aspas = re.findall(keywords_pattern, str(ementa))

    # Extrair termos-chave
    keywords_texto = [termo for termo in termos_chave if termo.lower() in str(ementa).lower()]

    # Combinar e remover duplicatas
    keywords = list(set(keywords_aspas + keywords_texto))

    return ', '.join(keywords) if keywords else 'Sem palavras-chave'

def classify_area_direito(ementa, keywords):
    """
    Classifica a área do direito com mais precisão
    """
    areas_mapeamento = {
        'Previdenciário': ['previdência', 'auxílio', 'aposentadoria', 'benefício', 'acidente de trabalho'],
        'Civil': ['ação civil', 'mandado de segurança', 'licitação', 'improbidade'],
        'Administrativo': ['servidor', 'município', 'estado', 'poder público']
    }

    ementa_lower = str(ementa).lower()
    keywords_lower = str(keywords).lower()

    for area, termos in areas_mapeamento.items():
        if any(termo in ementa_lower or termo in keywords_lower for termo in termos):
            return area

    return 'Não Classificado'

def processar_csv(input_file, output_file):
    """
    Processa o CSV original e gera novo CSV estruturado
    """
    # Ler o CSV original
    df = pd.read_csv(input_file, encoding='utf-8')

    # Criar novo DataFrame estruturado
    novo_df = pd.DataFrame({
        'numero_processo': df['numero_processo'],
        'tipo_decisao': df['tipo_decisao'].apply(mapear_tipo_decisao),
        'relator': df['relator'],
        'orgao_julgador': df['relator'].apply(extrair_orgao_julgador),
        'data_julgamento': df['data_julgamento'],
        'ementa_resumida': df['ementa'],
        'url_documento': df['url']
    })

    # Extrair keywords
    novo_df['keywords'] = novo_df['ementa_resumida'].apply(extract_keywords)

    # Classificar área do direito
    novo_df['area_direito'] = novo_df.apply(
        lambda row: classify_area_direito(row['ementa_resumida'], row['keywords']),
        axis=1
    )

    # Adicionar informações de página
    novo_df['pagina'] = df['pagina']
    novo_df['posicao_na_pagina'] = df['posicao_na_pagina']

    # Salvar novo CSV
    novo_df.to_csv(output_file, index=False, encoding='utf-8')

    print(f"\nCSV processado e salvo em {output_file}")
    print("\nVisão geral do novo DataFrame:")
    print(novo_df.head())
    print("\nContagem por área de direito:")
    print(novo_df['area_direito'].value_counts())
    print("\nContagem por tipo de decisão:")
    print(novo_df['tipo_decisao'].value_counts())

    return novo_df

# Caminho do diretório e arquivo
diretorio = '/content/extracao_tjpr_20250314_110949'
arquivo_input = '/content/extracao_tjpr_20250314_110949/resultados_parcial_ate_pagina_10.csv'
arquivo_output = '/content/extracao_tjpr_20250314_110949/resultados_estruturado_v2.csv'

# Processar o CSV
processar_csv(arquivo_input, arquivo_output)
```

```python
import pandas as pd
import re

def comparar_ementas(input_file):
    """
    Compara as ementas do CSV com o HTML original
    """
    # Ler o CSV
    df_csv = pd.read_csv(input_file, encoding='utf-8')

    # Ler o HTML original
    with open('/content/extracao_tjpr_20250314_110949/pagina_resultados_1.html', 'r', encoding='utf-8') as f:
        html_content = f.read()

    print("Comparação de Ementas:")
    print("-" * 50)

    # Extrair ementas do HTML
    ementa_padrao_html = re.findall(r'<div id="ementa\d+">(.*?)</div>', html_content, re.DOTALL)

    print(f"Número de ementas no CSV: {len(df_csv)}")
    print(f"Número de ementas no HTML: {len(ementa_padrao_html)}")

    # Comparar as primeiras 5 ementas
    print("\nComparação das primeiras 5 ementas:")
    for i in range(min(5, len(df_csv), len(ementa_padrao_html))):
        print(f"\nEmenta {i+1}:")
        print("CSV:  ", df_csv['ementa'].iloc[i][:200] + "...")
        print("HTML: ", ementa_padrao_html[i][:200] + "...")

        # Verificar similaridade
        similaridade = calcular_similaridade(df_csv['ementa'].iloc[i], ementa_padrao_html[i])
        print(f"Similaridade: {similaridade:.2f}%")

def calcular_similaridade(texto1, texto2):
    """
    Calcula a similaridade entre dois textos
    """
    # Remover tags HTML e caracteres especiais
    texto1_limpo = re.sub(r'<.*?>', '', texto1)
    texto2_limpo = re.sub(r'<.*?>', '', texto2)

    # Normalizar texto
    texto1_limpo = texto1_limpo.lower().strip()
    texto2_limpo = texto2_limpo.lower().strip()

    # Calcular similaridade de Jaccard
    set1 = set(texto1_limpo.split())
    set2 = set(texto2_limpo.split())

    intersecao = len(set1.intersection(set2))
    uniao = len(set1.union(set2))

    # Retornar porcentagem de similaridade
    return (intersecao / uniao) * 100 if uniao > 0 else 0

# Caminho do arquivo CSV
arquivo_input = '/content/extracao_tjpr_20250314_110949/resultados_parcial_ate_pagina_10.csv'

# Executar comparação
comparar_ementas(arquivo_input)
```
